## 1Ô∏è‚É£ Escape Characters (Fundamental Level)

An **escape character** is a backslash (`\`) used to tell Python:

> "The next character is part of the text, not part of the code."

It helps when a string contains characters that would normally end or break the string.

### üîπ The Problem

If we write:

```python
text = 'hello an'sh'
```

Python throws an error because:

- It thinks the string ends after `an`
- It gets confused by the second single quote

### üîπ Solution 1: Use Backslash (`\`)

````python
text = 'hello an\'sh'
print(text)
````

`\'` tells Python to treat the single quote as normal text.

### üîπ Solution 2: Use Double Quotes Outside

````python
text = "hello an'sh"
print(text)
````

No error occurs because:

- Outer quotes are double quotes (`"`)
- Inner quote is a single quote (`'`)

### üîπ Other Basic Escape Characters

````python
# New line
print("Hello\nWorld")

# Tab space
print("Name:\tStudent")
````

### üîπ Summary Table

| Escape | Meaning |
|--------|----------|
| `\'`   | Single quote |
| `\"`   | Double quote |
| `\n`   | New line |
| `\t`   | Tab |

## 2Ô∏è‚É£ String Literal Concatenation (Fundamental Level)

In Python, when multiple **string literals** are written next to each other, Python automatically joins them.

### üîπ Example Used by Instructor

````python
print('Hello ' 'Python Learner' '. This is a sample message.')
````

Python automatically combines the strings into:

```
Hello Python Learner. This is a sample message.
```

No `+` operator is required.

### üîπ Equivalent Version Using `+`

````python
print('Hello ' + 'Python Learner' + '. This is a sample message.')
````

Both approaches produce the same result.

### üîπ Important Difference

Automatic joining works only with **string literals written directly in the code**.

It does NOT work with variables.

````python
a = "Hello "
b = "World"

# Correct way with variables
print(a + b)
````

## ‚úÖ Fundamental Takeaways

- Escape characters prevent syntax errors.
- They allow special characters inside strings.
- Python automatically joins adjacent string literals.
- Variables require `+` for concatenation.
