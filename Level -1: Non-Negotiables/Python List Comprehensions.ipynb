# Nested List Comprehensions (Fundamentals)

This notebook explains **nested list comprehensions** and shows:
- Which `for` runs first (outer vs inner)
- How `if` conditions change the execution flow
- Filtering vs conditional expression (`if/else` inside output)

## 1) Printing with List Comprehension (Not Recommended, but possible)

### 1A) Print each item
```python
[print("  item:", item) for sub in nested for item in sub]
```

## 2) Flattening a Nested List (Most Common Real Use)

### List Comprehension
```python
flat = [item for sub in nested for item in sub]
```

### âœ… Which `for` runs first?
ðŸ‘‰ The **leftmost `for` runs first**.

So it behaves like:
```python
flat = []
for sub in nested:        # outer loop runs first
    for item in sub:      # inner loop runs second
        flat.append(item)
```


````python
flat = [item for sub in nested for item in sub]
print("flat =", flat)
````

## 3) Adding `if` Conditions

### Case 1 â€” `if` after inner loop
```python
flat_even = [item for sub in nested for item in sub if item % 2 == 0]
```

Equivalent to:
```python
flat_even = []
for sub in nested:
    for item in sub:
        if item % 2 == 0:
            flat_even.append(item)
```


````python
flat_even = [item for sub in nested for item in sub if item % 2 == 0]
print("flat_even =", flat_even)
````

### Case 2 â€” `if` after outer loop
```python
filtered = [item for sub in nested if len(sub) > 2 for item in sub]
```

Equivalent to:
```python
filtered = []
for sub in nested:
    if len(sub) > 2:
        for item in sub:
            filtered.append(item)
```

âœ… Here, the `if` decides whether we process the entire sub-list.


````python
filtered = [item for sub in nested if len(sub) > 2 for item in sub]
print("filtered =", filtered)
````

### Case 3 â€” Both outer and inner conditions

```python
result = [
    item
    for sub in nested
    if len(sub) > 1
    for item in sub
    if item % 2 == 0
]
```


````python
result = [
    item
    for sub in nested
    if len(sub) > 1
    for item in sub
    if item % 2 == 0
]
print("result =", result)
````

## 4) Conditional Expression (`if/else` inside output)

This returns a label for **every item**, so output length stays the same.

```python
labels = [
    "even" if item % 2 == 0 else "odd"
    for sub in nested
    for item in sub
]
```


````python
labels = [
    "even" if item % 2 == 0 else "odd"
    for sub in nested
    for item in sub
]
print("labels =", labels)
````

## 5) Combine Filtering + Conditional Expression (More Advanced)

```python
advanced = [
    "big_even" if item > 3 else "small_even"
    for sub in nested
    if len(sub) > 1
    for item in sub
    if item % 2 == 0
]
```

Meaning:
- Skip sub-lists of length 1
- Keep only even items
- Label the remaining even items based on size


````python
advanced = [
    "big_even" if item > 3 else "small_even"
    for sub in nested
    if len(sub) > 1
    for item in sub
    if item % 2 == 0
]
print("advanced =", advanced)
````

## âœ… Master Rule

Python reads comprehensions **top to bottom** just like nested loops.

General form:
```python
[ expression
  for outer in outer_iterable
  if outer_condition
  for inner in inner_iterable
  if inner_condition ]
```

Equivalent loops:
```python
for outer in outer_iterable:
    if outer_condition:
        for inner in inner_iterable:
            if inner_condition:
                expression
```
